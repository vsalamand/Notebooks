```python
"""Fetch unstructured text data from a given URL."""

from trafilatura import fetch_url, extract
import trafilatura 


def scrape_generic_text(url: str):
    downloaded = fetch_url(url)
    # to get the main text of a page
    if downloaded is not None:
        result = extract(downloaded, include_comments=False)
        text = story_parser(result)
        if text is not None:
            df = pd.DataFrame(text[1:], columns=['text'])
            return df
        else:
            pass

def story_parser(result):
    # getting all the paragraphs
    text = []
    try:
        sentences = sent_tokenize(result, language='french')
        for sentence in sentences:
            if (sentence.replace("\n","* ").replace("– ","* ").replace("- ","* ").replace("• ","* ").replace("• ","* ").count('*') > 2):
                break
            else:
                text.append(sentence.replace('\n', ' ').replace('\r', ' ').replace('\xa0', ' ').strip())

        return text
    except:
        return None

    
    
```

```python
# load recipe urls from csv files
import glob
import os

folder_name = '/Users/vincentsalamand/Documents/Clubmama/Recipes/'
file_type = 'csv'
seperator =','
dataframe = pd.concat([pd.read_csv(f, sep=seperator) for f in glob.glob(folder_name + "/*."+file_type)],ignore_index=True)



```

```python
frames = []
for link in dataframe.link:
    print(link)
    frames.append(scrape_generic_text(link))
    
```

```python
# Create labeled dataset
various_text = pd.concat(frames)
various_text = various_text.reset_index()
various_text = various_text.drop(columns=['index'])
print(len(various_text))
various_text.tail()

```
