```python

categories = []

for main_cat in main_categories:
    parent_shelter = main_cat['name']
    driver.get(main_cat['url'])
    sleep(1)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    results = soup.select('.category-navigation-item')
    for result in results:
        if parent_shelter == result.find("div", class_="f-collapse-item__title").text.strip():
            for li in result.find_all("li")[1:]:
                child_shelter = li.text.strip()
                url = 'https://www.franprix.fr' + li.find("a")["href"]
                categories.append({"parent_shelter": parent_shelter, 
                                   "child_shelter": child_shelter,
                                   "url": url
                                  })
                
len(categories)
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.70 Safari/537.36'}
product_pages = []

for category in categories:

    url = category['url']
    driver.get(url)
    sleep(1)
    
    for i in range(15):
        try:
            actions = ActionChains(driver)
            actions.move_to_element(driver.find_elements_by_class_name("product-list-item")[-1]).perform()
        except:
            pass  
    
    try:
        soup = BeautifulSoup(driver.page_source, 'lxml')
        results = soup.select('.product-list')[0]

        product_pages.append({"url": url, 
                              "shelter_parent": category['parent_shelter'],
                              "child_shelter": category['child_shelter'],
                              "soup": results
                              })
    except:
        print(url)
        
        
len(product_pages)    
    
```

```python
data = pd.DataFrame(product_pages)
data.tail()
```

```python
products = []
bad_products = []

for page in product_pages:
    shelter_parent = page['shelter_parent']
    shelter_child = page['child_shelter']   
    
    for product in page['soup'].find_all("div", class_="product-list-item"):
        product_name = product.find("h2", class_="product-item-name").text.strip()
        product_link = "https://www.franprix.fr" + product.find("div", class_="product-item").find('a')['href']
        try:
            product_image = product.find('div', class_='product-item-image').find('img')['src']
        except:
            bad_products.append(product_name)
        product_ean = ''.join([n for n in product_image if n.isdigit()])[:13]
        product_price = product.find('span', class_='product-item-price').text
        store_name = "Franprix"

        more_infos = product.find('span', class_='product-item-more').find_all('span')
        if len(more_infos) == 1:
            quantity = more_infos[0].text.strip()
        elif len(more_infos) == 2:
            brand = more_infos[0].text.strip()[:-2]
            quantity = more_infos[1].text.strip()    
        else:
            print(more_infos)
            bad_products.append(product_name)

        products.append({"shelter_parent": shelter_parent,
                         "shelter_child": shelter_child,
                         "brand": brand,  
                         "description1": product_name,                             
                         "description": product_name,
                         "price": product_price,
                         #"price_per_unit": product_price_per_unit,
                         "quantity": quantity,
                         #"promo_price": promo_price,
                         "store": store_name,
                         "image_url": product_image,                    
                         "url": product_link,
                         "ean": product_ean
                        })
print(len(products))
print(len(bad_products))
```

```python
# find missing eans

import re

product_data = pd.DataFrame(products)

if len(product_data[product_data['ean'] == '']) > 0:
    for index, product in product_data.iterrows(): 
        if product.ean == '':
            driver.get(product.url)
            soup = BeautifulSoup(driver.page_source, 'lxml')        
            try:
                ean = re.findall(r"\d{13}", soup.text)[0]
                product_data.loc[index, 'ean'] = ean
                print(ean)
            except:
                print(product.url)
                pass

```

```python
# drop product without ean
print(len(product_data[product_data['ean'] == '']))
product_data['ean'].replace('', np.nan, inplace=True)
product_data = product_data.dropna(subset=['ean'])
print(len(product_data[product_data['ean'] == '']))
```
