```python
# spaCy based imports
import spacy
from spacy.lang.fr.stop_words import STOP_WORDS
from spacy.lang.fr import French
!python -m spacy download fr_core_news_md
```

```python
nlp = spacy.load('fr_core_news_md')
```

```python
import re
import unidecode

# Convert to list
data = posts.post.astype(str).tolist()

# Convert to unicode and remove accents
data = [unidecode.unidecode(item) for item in data]

# Remove Emails
data = [re.sub('\S*@\S*\s?', "", str(sent)) for sent in data]

# Remove new line characters
data = [re.sub('\s+', " ", str(sent)) for sent in data]

# Remove distracting single quotes
data = [re.sub("\'", "", str(sent)) for sent in data]

# Remove return line characters
data = [re.sub('\n', " ", str(sent)) for sent in data]

# Remove non alphabetical characters
data = [re.sub(r'[^A-Za-z]', " ", str(sent)) for sent in data]

# Lower all characters
data = [str(sent).lower() for sent in data]
```

```python
docs = nlp.tokenizer.pipe(data)
texts = []
for doc in docs:
    doc = [token for token in doc]
    doc = [token for token in doc if not token.is_stop]
    doc = [token for token in doc if token.whitespace_]
    doc = [token.lemma_ for token in doc]
    texts.append(doc)
```

```python
# Gensim
import gensim
from gensim import corpora, models
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
```

```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(texts, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[texts], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

# See trigram example
print(trigram_mod[bigram_mod[texts[0]]])

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]
```

```python
# Form Bigrams
texts = make_bigrams(texts)
```

```python
# Form Bigrams
texts = make_bigrams(texts)

# Create Dictionary
id2word = corpora.Dictionary(texts)

# Create Corpus
texts = texts

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:1])
```

```python
#TF-IDF¶
tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```python
# Human readable format of corpus (term-frequency)
[[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]]
```

```python
# Running LDA using Bag of Words (corpus)
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=10, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
from pprint import pprint

# Print the Keyword in the n topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
# Running LDA using TF-IDF
lda_model_tfidf = gensim.models.LdaMulticore(corpus_tfidf, 
                                             id2word=id2word,
                                             num_topics=20, 
                                             passes=10, 
                                             workers=4)
```

```python
from pprint import pprint

# Print the Keyword in the n topics
pprint(lda_model_tfidf.print_topics())
doc_lda = lda_model_tfidf[corpus]
```

```python
# Compute Perplexity
print('\nPerplexity: ', lda_model_tfidf.log_perplexity(corpus))  # a measure of how good the model is. lower the better.

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model_tfidf, texts=texts, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

```python
#Testing model on unseen document
unseen_document = texts[1]
bow_vector = id2word.doc2bow(texts[1])

for index, score in sorted(lda_model_tfidf[bow_vector], key=lambda tup: -1*tup[1]):
    print("Score: {}\t Topic: {}".format(score, lda_model_tfidf.print_topic(index, 5)))
```
