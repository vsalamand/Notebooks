```python
import pandas as pd 
import numpy as np
import re
```

```python
items = pd.read_csv('/Users/vincentsalamand/Documents/Clubmama/db_exports/item_export.csv')
items = items.drop(columns=['created_at', 'updated_at', 'list_item_id'])
items = items.rename(columns={"id": "item_id", "name": "item_name"})

foods = pd.read_csv('/Users/vincentsalamand/Documents/Clubmama/db_exports/food_export.csv')
units = pd.read_csv('/Users/vincentsalamand/Documents/Clubmama/db_exports/unit_export.csv')
recipes = pd.read_csv('/Users/vincentsalamand/Documents/Clubmama/db_exports/recipes_export.csv')
recipe_lists = pd.read_csv('/Users/vincentsalamand/Documents/Clubmama/db_exports/recipe_lists_export.csv')

```

```python
# Convert ids in items df to integrer in order to join with related tables
items.food_id = items.food_id.fillna(0)
items.food_id = items.food_id.astype(int)

items.recipe_id = items.recipe_id.fillna(0)
items.recipe_id = items.recipe_id.astype(int)


items.unit_id = items.unit_id.fillna(0)
items.unit_id = items.unit_id.astype(int)
```

```python
#map items with food, unit, and recipes names

item_with_labels = items.merge(foods[['id','name']], how='left',left_on=['food_id'], right_on=['id'])
item_with_labels = item_with_labels.merge(recipes[['id','title']], how='left',left_on=['recipe_id'], right_on=['id'])
item_with_labels = item_with_labels.merge(units[['id','name']], how='left', left_on=['unit_id'], right_on=['id'])

item_with_labels = item_with_labels.drop(columns=['id_y', 'id_x', 'id'])
item_with_labels = item_with_labels.rename(columns={"name_x": "food_name", "title": "recipe_name", "name_y": "unit_name"})

item_with_labels = item_with_labels.dropna(subset = ['item_name'])
```
